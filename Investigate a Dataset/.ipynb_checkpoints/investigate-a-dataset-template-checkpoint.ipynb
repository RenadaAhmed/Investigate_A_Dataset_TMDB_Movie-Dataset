### General Properties

```python
df_tmbd = pd.read_csv("tmdb-movies.csv")
df_tmbd.head()
```

some Movies has zero revenus and budget so they irrelevent as they can be considered as nan values  

```python
df_tmbd.isnull().sum()
```

```python
df_tmbd.duplicated().sum()
```

```python
df_tmbd.drop(['id','imdb_id','homepage','tagline','keywords','overview'],axis=1 ,inplace=True)
df_tmbd.head()
```

```python
df_tmbd.drop_duplicates(inplace=True)
```

```python
df_tmbd.duplicated().sum()
```

remove movies with zero revenues or budgets

```python
df_nobud = df_tmbd.query('budget == 0 & budget_adj ==0')
df_tmbd.drop(df_nobud.index,inplace=True)
```

```python
df_tmbd.shape
```

```python
df_tmbd['release_date'] = pd.to_datetime(df_tmbd['release_date'], infer_datetime_format=True)
```

```python
df_tmbd.release_year=df_tmbd.release_year.astype(str)
```

```python
df_tmbd.dropna(inplace=True)
```

```python
df_tmbd.info()
```

```python
df_tmbd.isnull().sum()
```

```python
df_tmbd.describe()
```

```python
df_tmbd.shape
```

## Exploratory Data Analysis

###  Question 1 (Which genres are most popular from year to year?)

```python
df_tmbd.columns
```

```python
for i, v in enumerate(df_tmbd.columns):
    print(i, v)
```

```python
df_gen = pd.DataFrame(data=df_tmbd.genres.str.split('|').tolist(),index=[df_tmbd.release_year,df_tmbd.popularity]).stack().reset_index([0,1])
df_gen.rename(mapper={0:'geners'},axis=1,inplace=True)
df_gen.head()
```

```python
df_gen.geners.nunique()
```

```python
df_gen.release_year=df_gen.release_year.astype(int)
df_gen.release_year = df_gen.release_year.apply(lambda x:(x //10)*10)
df_gen_edit = pd.DataFrame(df_gen.groupby(['release_year','geners'],as_index=False).popularity.sum())
df_gen_edit.head(15)
```

```python
df_gen_edit.tail(30)
```

```python
df_gen_clean = pd.DataFrame(df_gen.groupby(['release_year','geners']).popularity.sum())
df_gen_clean.head(20)
```

```python
df_gen_clean.unstack().plot(figsize=(12,15),linewidth=2.5);   #kind='barh',
plt.title("Most Popular Genre Over Year To Year",fontsize=12);
```

```python
fig, ax = plt.subplots(3,2,figsize = (12,9))
df_gen_clean.query("geners == 'Drama'").plot(kind='bar',label ="Drama",ax = ax[0][0]);
df_gen_clean.query("geners == 'Action'").plot(kind='bar',label ="Action",ax = ax[0][1]);
df_gen_clean.query("geners == 'Adventure'").plot(kind='bar',label ="Adventure",ax = ax[1][0]);
df_gen_clean.query("geners == 'Thriller'").plot(kind='bar',label ="Thriller",ax = ax[1][1]);
df_gen_clean.query("geners == 'Comedy'").plot(kind='bar',label ="Comedy",ax = ax[2][0]);
df_gen_clean.query("geners == 'Science Fiction'").plot(kind='bar',label ="Science Fiction",ax = ax[2][1]);

fig.suptitle('Genre Popularity Over Year To Year',fontsize = 13);
```

###  Question 2  (What kinds of properties are associated with movies that have high revenues ?)

```python
#df_rev = pd.DataFrame(df_tmbd['revenue_adj'].sort_values(ascending=False))
df_rev = df_tmbd.set_index('revenue', append=True).sort_index(level=1,ascending=False).reset_index(level=1)
df_rev.drop(['budget_adj','revenue_adj'],axis=1 ,inplace=True)
df_rev.head(3)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_rev.corr(method='pearson', min_periods=1),annot=True ,cmap="YlGnBu")
```

```python
X = df_rev.drop('revenue', axis=1)
y = df_rev['revenue']
X.corrwith(y).plot(kind='bar', grid=False, figsize=(7, 8), title="Correlation with the Revenue");
```

```python
fig, axes = plt.subplots(3,2,figsize = (10,9))
axes[2][1].set_visible(False)

fig.suptitle("Revenus relation with other properities",fontsize=15)
sns.regplot(x=df_rev['revenue'], y=df_rev['budget'],ax=axes[0][0])
sns.regplot(x=df_rev['revenue'], y=df_rev['popularity'],ax=axes[0][1])
sns.regplot(x=df_rev['revenue'], y=df_rev['vote_average'],ax=axes[1][0])
sns.regplot(x=df_rev['revenue'], y=df_rev['runtime'],ax=axes[1][1])
sns.regplot(x=df_rev['revenue'], y=df_rev['vote_count'],ax=axes[2][0])
```

###  Question 3  (Which movie has the highest profit ?)
